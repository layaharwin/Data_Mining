```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Reading the dataset
df= pd.read_csv("TikTok_songs_2019.csv")
#Gist of the dataset
df.head()
```

```python
# Consice summary of data frame
# Displays the type, count and name of the attributes
df.info()
```

```python
#Statistical description of the dataset
df.describe()
```

```python
# Includes categorical variable 
df.describe(include='all')
```

```python
#See the types
df.dtypes
```

```python
# Understanding the dataset more

# Number of enteries 
print("Number of Enteries:",df.shape)
# We have 198917 rows and 14 columns 

# Total number of products & unique values of the columns 
print("*****************")
print("Total number of products & unique values of the columns:\n",df.nunique())

# Considering mode attribute. It has values 0 and 1.
# Count when mode is 0 and mode is 1 
print("*****************")
print("Count when mode=1:",df[df['mode'] == 1]['duration_ms'].count())
print("Count whne mode=0:",df[df['mode'] == 0]['duration_ms'].count())

# Split the dataset into two parts based on the mode values 
print("*****************")
df_mode0 = df[df['mode'] == 0]
df_mode1 = df[df['mode'] == 1]
print(df_mode0.describe())
print("*****************")
print(df_mode1.describe())
```

```python
%matplotlib inline
#Check out data using  plots
#Checking out Mode attribute
df['mode'].value_counts().plot.bar(title="Mode count:")
```

```python
# Plotting histogram
# Checking out attribute: liveness
sns.set_style('white')
sns.set_context("paper", font_scale = 2)
sns.displot(data=df, x="liveness", kind="hist", bins = 100, aspect = 1.5)
```

```python
#Histogram plot for numerical attributes
#Plotting 'loudness','track_pop','tempo','duration_ms' in one canvas
cols = ['loudness','track_pop','tempo','duration_ms']
fig, axis = plt.subplots(len(cols), figsize=(16,12))
#iterating through each attribute and representing their values through histogram
for i, col_val in enumerate(cols):       
    sns.histplot(df[col_val], ax=axis[i],kde=True, stat="density", linewidth=0) #hist=True, ax=axis[i])
    axis[i].set_title('Freq dist '+col_val, fontsize=12)
    axis[i].set_xlabel(col_val, fontsize=12)
    axis[i].set_ylabel('Count', fontsize=12)
plt.show()
```

```python
#Finging outliers
cols1 = ['loudness','track_pop','tempo','duration_ms']
#fig, axis = plt.subplots(len(cols), figsize=(16,12))
fig, axis = plt.subplots(len(cols1), figsize=(5,18))
for i, cols_name in enumerate(cols1):    
    sns.boxplot(y=df[cols_name], ax=axis[i])
    axis[i].set_title('Box plot - '+cols_name, fontsize=8)
    axis[i].set_xlabel(cols_name, fontsize=8)
plt.show()
#here we can see that loudness','track_pop' and'duration_ms have outliers. Outliers are points present outside the box-plot
```

```python
#Bivariate plots : to plot multiple pairwise bivariate distributions 
sns.pairplot(df)
```

```python
# Missind Data
# Approach 2:
# Null Check
df.isnull().sum() # here we are checking the attributes that have null/missing values. Here also we can see that none of the attribute have missing data hence 0. 
```

```python
#Correlation between attributes
print('Correlation(Pearson):')
# By default we are doing pearson: standard correlation coefficient 
df.corr()
# here we can see correlation between attributes
# if correlation is negative: attributes are inversely proportional
# if correlation is positive: attributes are proportional
```

```python
#Correlation between attributes
# kendall: Kendall Tau correlation coefficient
print('Correlation(Kendall):')
df.corr(method="kendall")
```

```python
#Correlation between attributes
#spearman: Spearman rank correlation
print('Correlation(Spearman):')
df.corr(method="spearman")
```
